```python
import numpy as np
import matplotlib.pyplot as plt

# ===== Blocked Environment =====
b_a_o = np.array([6.3])
b_a_o_t = np.cumsum(np.array([4.2]))

b_a_e = np.array([6.3])
b_a_e_t = np.cumsum(np.array([2.5]))

b_a_m = np.array([6.5])  # Manhattan
b_a_m_t = np.cumsum(np.array([3.8]))

b_a_c = np.array([6.4])  # Chebyshev
b_a_c_t = np.cumsum(np.array([3.0]))

b_an_o = np.array([10.2, 6.3])
b_an_o_t = np.cumsum(np.array([1.16, 1.8]))

b_an_e = np.array([11.1, 7.3, 6.3])
b_an_e_t = np.cumsum(np.array([1.7, 1.4, 1.1]))

b_an_m = np.array([10.5, 6.5])  # Manhattan ANA*
b_an_m_t = np.cumsum(np.array([1.3, 1.9]))

b_an_c = np.array([10.4, 6.4])  # Chebyshev ANA*
b_an_c_t = np.cumsum(np.array([1.2, 1.5]))

# Plot Blocked Environment
def plot_env(title, a_o, a_o_t, a_e, a_e_t, a_m, a_m_t, a_c, a_c_t, an_o, an_o_t, an_e, an_e_t, an_m, an_m_t, an_c, an_c_t):
    tmax = np.max([a_o_t[-1], a_e_t[-1], a_m_t[-1], a_c_t[-1], an_o_t[-1], an_e_t[-1], an_m_t[-1], an_c_t[-1]])
    cmax = np.max(np.concatenate([a_o, a_e, a_m, a_c, an_o, an_e, an_m, an_c]))

    plt.figure(figsize=(10, 6))

    # A* Results
    plt.vlines(a_o_t, a_o, cmax + 5, label='Octile, A*', colors='k')
    plt.vlines(a_e_t, a_e, cmax + 5, label='Euclidean, A*', colors='b')
    plt.vlines(a_m_t, a_m, cmax + 5, label='Manhattan, A*', colors='c')
    plt.vlines(a_c_t, a_c, cmax + 5, label='Chebyshev, A*', colors='m')

    # ANA* Results
    for m, i in enumerate(an_o):
        if m == 0:
            plt.vlines(an_o_t[m], cmax + 5, i, colors='r')
        if m == len(an_o) - 1:
            plt.hlines(i, an_o_t[m], tmax, colors='r', label='Octile, ANA*')
        if 0 <= m < len(an_o) - 1:
            plt.hlines(i, an_o_t[m], an_o_t[m + 1], colors='r')

    for m, i in enumerate(an_e):
        if m == 0:
            plt.vlines(an_e_t[m], cmax + 5, i, colors='g')
        if m == len(an_e) - 1:
            plt.hlines(i, an_e_t[m], tmax, colors='g', label='Euclidean, ANA*')

    for m, i in enumerate(an_m):
        if m == 0:
            plt.vlines(an_m_t[m], cmax + 5, i, colors='c')
        if m == len(an_m) - 1:
            plt.hlines(i, an_m_t[m], tmax, colors='c', label='Manhattan, ANA*')

    for m, i in enumerate(an_c):
        if m == 0:
            plt.vlines(an_c_t[m], cmax + 5, i, colors='m')
        if m == len(an_c) - 1:
            plt.hlines(i, an_c_t[m], tmax, colors='m', label='Chebyshev, ANA*')

    plt.legend(loc='lower left')
    plt.xlabel('Time (s)')
    plt.ylabel('Costs')
    plt.title(title)
    plt.xlim(0, tmax + 1)
    plt.ylim(0, cmax + 2)
    plt.show()

# ===== MoveSev Environment =====
m_a_o = np.array([6])
m_a_o_t = np.cumsum(np.array([49.4]))

m_a_e = np.array([5.9])
m_a_e_t = np.cumsum(np.array([65.8]))

m_a_m = np.array([6.5])  # Manhattan
m_a_m_t = np.cumsum(np.array([70.2]))

m_a_c = np.array([6.2])  # Chebyshev
m_a_c_t = np.cumsum(np.array([60.3]))

m_an_o = np.array([7.1, 6.3])
m_an_o_t = np.cumsum(np.array([9.2, 75.2]))

m_an_e = np.array([12.9, 6.9, 6.2])
m_an_e_t = np.cumsum(np.array([21.9, 18.3, 15.7]))

m_an_m = np.array([10.5, 6.5])
m_an_m_t = np.cumsum(np.array([8.5, 60.1]))

m_an_c = np.array([10.4, 6.4])
m_an_c_t = np.cumsum(np.array([7.3, 55.5]))

# ===== PR2Doorway Environment =====
d_a_o = np.array([10.2])
d_a_o_t = np.cumsum(np.array([23.2]))

d_a_e = np.array([10.3])
d_a_e_t = np.cumsum(np.array([29.5]))

d_a_m = np.array([11.0])  # Manhattan
d_a_m_t = np.cumsum(np.array([28.5]))

d_a_c = np.array([10.8])  # Chebyshev
d_a_c_t = np.cumsum(np.array([25.7]))

d_an_o = np.array([10.4])
d_an_o_t = np.cumsum(np.array([3.7]))

d_an_e = np.array([10.4])
d_an_e_t = np.cumsum(np.array([7.6]))

d_an_m = np.array([11.1])
d_an_m_t = np.cumsum(np.array([6.5]))

d_an_c = np.array([10.9])
d_an_c_t = np.cumsum(np.array([5.7]))

# ===== Plots =====
plot_env("Env: Blocked", b_a_o, b_a_o_t, b_a_e, b_a_e_t, b_a_m, b_a_m_t, b_a_c, b_a_c_t, 
         b_an_o, b_an_o_t, b_an_e, b_an_e_t, b_an_m, b_an_m_t, b_an_c, b_an_c_t)

plot_env("Env: MoveSev", m_a_o, m_a_o_t, m_a_e, m_a_e_t, m_a_m, m_a_m_t, m_a_c, m_a_c_t, 
         m_an_o, m_an_o_t, m_an_e, m_an_e_t, m_an_m, m_an_m_t, m_an_c, m_an_c_t)

plot_env("Env: PR2Doorway", d_a_o, d_a_o_t, d_a_e, d_a_e_t, d_a_m, d_a_m_t, d_a_c, d_a_c_t, 
         d_an_o, d_an_o_t, d_an_e, d_an_e_t, d_an_m, d_an_m_t, d_an_c, d_an_c_t)
```
